```python
files = []
file_label_dict = {}
files_src = "../inputdata/US_states_max100.csv"
with open(files_src, encoding="utf-8",newline="") as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        files.append(row[0])
        file_label_dict[row[0]] = row[1]
imagelist = [file for file in file_label_dict.keys() if file_label_dict[file] == "Arizona"]
```
